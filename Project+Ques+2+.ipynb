## Basic information about President Franklin d Roosevelt  speech

```python
num_chars = len(inaugural.raw('1941-Roosevelt.txt'))
num_words = len(inaugural.words('1941-Roosevelt.txt'))
num_sents = len(inaugural.sents('1941-Roosevelt.txt'))

print(" The numbers of characters in Pres Roosevelt 1941 speech is:",num_chars)
print(" The numbers of words in Pres Roosevelt 1941 speechis:",num_words)
print(" The numbers of sentences in Pres Roosevelt 1941 speech is:",num_sents)
```

## Processing the Text Or Cleaning the Document

```python
## Converting to Lower Case
raw = inaugural.raw('1941-Roosevelt.txt').lower()
raw
```

```python
## Cleaning the special characters (/@$: etc)
## from Regular Expressions package
## re.sub() function for string substitution using regular expressions
import re
raw = re.sub("[^a-zA-Z]"," ",str(raw))
```

```python
raw
```

```python
## Tokenization -  Splitting the text files into words
tokens = nltk.word_tokenize(raw)
print(tokens)
```

```python
len(tokens)
```

```python
## Stop words -  Removing words that are not useful or Removing meaningless words 
from nltk.corpus import stopwords
en_stop = stopwords.words('english')
stopped_tokens = [i for i in tokens if not i in en_stop]
```

```python
len(stopped_tokens)
```

```python
stopped_tokens.sort()
```

```python
print(stopped_tokens)
```

```python
## Stemming Using Lemmatizer -  Stem the words to its root word
from nltk import WordNetLemmatizer
lt = nltk.WordNetLemmatizer()
texts = [lt.lemmatize(i) for i in stopped_tokens]
```

```python
## Top 3 frequency occuring words
roosevelt_3 = nltk.FreqDist(texts).most_common(3)
roosevelt_3
```

```python
## Plotting the top 30 words occuring frequently
plt.figure(figsize=(12, 8))
nltk.FreqDist(texts).plot(30);
```

### Dispersion Plot
Lexical dispersion is a measure of how frequently a word appears across the parts of a corpus.

```python

from nltk.draw.dispersion import dispersion_plot
plt.figure(figsize=(10,8))
words_used=["spirit","people","democracy", "america",'faith','freedom']
dispersion_plot(texts, words_used, ignore_case=True, title='Dispersion Plot')
```

```python
no_order = list(set(texts))
no_order.sort()
print(no_order)
len(no_order)
```

## Creating a WordCloud

```python
from wordcloud import WordCloud, STOPWORDS

new_words = set(STOPWORDS)
new_words.update(['u','must', 'let','midst','like','come','may','go','enough','even','many','still','early','every','within'])

plt.subplots(figsize=(10,10))
wordcloud = WordCloud(background_color='black',max_words=75,
                            width=1200,stopwords=new_words ,
                            height=1000,colormap='tab20c',collocations=False).generate(" ".join(texts))

plt.title("MOST USED WORDS",fontsize=20,color='blue')
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
